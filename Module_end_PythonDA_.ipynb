Exploratory Data Analysis (EDA) and Visualization
Campaign Performance :

```python
# Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI.
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(marketing_df["Acquisition_Cost"], marketing_df["ROI"],color="Blue",s=3)
plt.xlabel("Acquisition_Cost")
plt.ylabel("ROI")
plt.title("Scatter plot of Acquisition_Cost vs ROI")
plt.show()
```

```python
# Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.
```

```python
avg_conversion = marketing_df.groupby(['Channel_Used', 'Campaign_Type'])['Conversion_Rate'].mean().unstack()
avg_conversion.plot(kind='bar', stacked=True)
plt.title('Average Conversion Rate by Channel and Campaign Type')
plt.xlabel('Channel Used')
plt.ylabel('Average Conversion Rate')
plt.legend(title='Campaign Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualize the distribution of Engagement_Score across different Campaign_Type using a box plot.
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.boxplot(data=marketing_df, x='Campaign_Type', y='Engagement_Score', palette='Set2')

plt.title('Distribution of Engagement Score across Campaign Types')
plt.xlabel('Campaign Type')
plt.ylabel('Engagement Score')
plt.show()
```

```python
# Analyze the average ROI by Company using a bar chart to compare the profitability of campaigns conducted by different companies.
```

```python
avg_roi = marketing_df.groupby('Company')['ROI'].mean().sort_values(ascending=False)
plt.bar(avg_roi.index, avg_roi.values)
plt.title('Average ROI by Company')
plt.xlabel('Company')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.show()
```

```python
# Examine the correlation between Engagement_Score and Conversion_Rate using a heatmap.
```

```python
sns.heatmap(marketing_df[['Engagement_Score', 'Conversion_Rate']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Customer Segmentation:

```python
# Create a count plot to visualize the distribution of Target_Audience.
```

```python
plt.figure(figsize=(10,8))
sns.countplot(data=marketing_df, x='Target_Audience')
plt.title('Distribution of Target Audience')
plt.xlabel('Target Audience')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify which Customer_Segment has the highest Conversion_Rate for each Language using a bar chart.
```

```python
highest_conv = (
    marketing_df.groupby(['Language', 'Customer_Segment'])['Conversion_Rate']
      .mean()
      .reset_index()
)
top_segments = highest_conv.loc[highest_conv.groupby('Language')['Conversion_Rate'].idxmax()]
plt.figure(figsize=(10,6))
plt.bar(top_segments['Language'], top_segments['Conversion_Rate'], color='lightcoral')
for i, v in enumerate(top_segments['Conversion_Rate']):
    plt.text(i, v + 0.01, top_segments.iloc[i]['Customer_Segment'], ha='center', fontsize=9)
plt.title('Highest Conversion Rate by Customer Segment for Each Language')
plt.xlabel('Language')
plt.ylabel('Conversion Rate')
plt.tight_layout()
plt.show()
```

```python
# Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used, using a box plot.
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Customer_Segment', y='Acquisition_Cost', hue='Channel_Used', data=marketing_df)
plt.title('Acquisition Cost across Customer Segments by Channel')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze average Conversion_Rate by Language using a bar chart to compare the effectiveness of campaigns conducted in different languages.
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Language', y='Conversion_Rate', data=marketing_df, estimator='mean', palette='Set2')
plt.title('Average Conversion Rate by Language')
plt.show()
```

Channel Effectiveness:

```python
# Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type, using a bar chart.
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Channel_Used', y='Engagement_Score', hue='Campaign_Type', data=marketing_df)
plt.title('Engagement Score by Channel and Campaign Type')
plt.xticks(rotation=45)
plt.show()
```

```python
# Show the distribution of total ROI across different Channels_Used using a pie chart.
```

```python
roi_by_channel = marketing_df.groupby('Channel_Used')['ROI'].sum()
plt.figure(figsize=(7,7))
plt.pie(roi_by_channel, labels=roi_by_channel.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Total ROI by Channel Used')
plt.show()
```

```python
# Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Impressions', y='Clicks', hue='Channel_Used', data=marketing_df)
plt.title('Relationship between Clicks and Impressions by Channel')
plt.show()
```

Time-Based Analysis:

```python
# Plot the distribution of Duration using a histogram.
```

```python
plt.figure(figsize=(8,6))
plt.hist(marketing_df['Duration'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Campaign Duration')
plt.xlabel('Duration (days)')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze how the overall Conversion_Rate has changed over Date for each Company using a line chart.
```

```python
marketing_df['Date'] = pd.to_datetime(marketing_df['Date'], dayfirst=True, errors='coerce')
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Conversion_Rate', hue='Company', data=marketing_df)
plt.title('Trend of Conversion Rate Over Time by Company')
plt.show()
```

```python
# Examine the trend of Engagement_Score over Date with a line chart.
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Engagement_Score', data=marketing_df)
plt.title('Trend of Engagement Score Over Time')
plt.show()
```

Geographic Analysis

```python
# Determine which location has the highest Acquisition_Cost using a bar chart.
```

```python
avg_acq = marketing_df.groupby('Location')['Acquisition_Cost'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
avg_acq.plot(kind='bar', color='lightgreen')
plt.title('Average Acquisition Cost by Location')
plt.ylabel('Acquisition Cost')
plt.show()
```

```python
# Visualize the Conversion_Rate by different Location, categorized by Target_Audience, using a bar chart.
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Location', y='Conversion_Rate', hue='Target_Audience', data=marketing_df)
plt.title('Conversion Rate by Location and Target Audience')
plt.xticks(rotation=45)
plt.show()
```

```python
# Illustrate the proportion of ROI by Location using a pie chart.
```

```python
roi_by_loc = marketing_df.groupby('Location')['ROI'].sum()
plt.figure(figsize=(7,7))
plt.pie(roi_by_loc, labels=roi_by_loc.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of ROI by Location')
plt.show()
```
